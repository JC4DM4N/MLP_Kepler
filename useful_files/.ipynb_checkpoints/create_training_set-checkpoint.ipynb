```python
def create_training_set(koi, system_ids, save=False):
    times  = []
    fluxes = []
    targets= []
    for system_id in system_ids:
        try:
            # this currently just picks out the first light curve of each system and only appends
            # if the shape is 1624
            a = clean(system_id, koi, lc_index=0, slc_or_llc='llc', verbose=False)
            a.remove_stellar_activity()
            if a.n == 1624:
                times.append(a.time)
                fluxes.append(a.flux)
                targets.append(a.target)
        except:
            pass
#             print('oh poop looks like we dont have data for that system\n')
    times   = np.array(times)
    fluxes  = np.array(fluxes)
    targets = np.array(targets)
    if save:
        np.savetxt('../data/train_test/times.csv', times, delimiter=',')
        np.savetxt('../data/train_test/fluxes.csv', fluxes, delimiter=',')
        np.savetxt('../data/train_test/targets.csv', targets, delimiter=',')
    return times, fluxes, targets
```

```python
times, fluxes, targets = create_training_set(koi, koi.index.values[:2500], save=False)
```
